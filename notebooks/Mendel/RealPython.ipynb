### Define a Function (def)

```python
# This is a function definition

def print_greeting(name):
    print(f"Hello, {name}!")
print_greeting("Alice")
```

### Def: Combine First + Last Name

```python
def make_full_name(first_name, last_name):
    """Combine first and last name into a full name."""
    full_name = first_name + " " + last_name
    return full_name
make_full_name("Real", "Python")
```

Alternative (f-function)

```python
def make_full_name(first_name, last_name):
    print(f"{first_name} {last_name}")
make_full_name("Real", "Python")
```

### Def: Print on Two Lines

```python
def print_profile(name, language):
    name_line = "Name: " + name
    language_line = "Language: " + language
    print(name_line)
    print(language_line)
print_profile("Jean Baptiste", "French")
```

Alternative (f-function)

```python
def print_profile(name, language):
    print(f"Name: {name}")
    print(f"Language: {language}")
print_profile("Real Python", "Python")
```

### for Loops

```python
# Test with a list of items. Tuples (in parentheses) work the same way, but they are immutable (cannot be changed).
a = ["foo", "bar", "baz"]
for item in a:
    print(item)
```

```python
# Test with a dictionary of items.
a = {"foo": 1, "bar": 2, "baz": 3}
for item in a:
    print(item)

print()

# Test with a dictionary of items. Return values.
a = {"foo": 1, "bar": 2, "baz": 3}
for item in a.values():
    print(item)

print()

# Test with a dictionary of items. Return values.
a = {"foo": 1, "bar": 2, "baz": 3}
for k, n in a.items():
    print(f"key={k}, num={n}")
```

```python
# Test examples with a string.
a = ("Hello World")
for item in a:
    print(item)
```

```python
# Test with a set of items. 
a = {"foo", "bar", "baz"}
for item in a:
    print(item)
```

```python
# Test with list and tuples.
points = [(1,4), (2,3), (3,2)]
    
for x, y in points:
        print(f"{x = } and {y = }")
```

#### Range

```python
a = range(5)
for i in a:
    print(i)

print()

a = range(0,5)
for i in a:
    print(i)

print()

a = range(0,5,2)
for i in a:
    print(i)
```

#### Break & Continue & Else

```python
a = ["brent", "taylor", "mitch", "jason", "buxby"]
for name in a:
    if "y" in name:
        break 
    print(name)

print()

a = ["brent", "taylor", "mitch", "jason", "buxby"]
for name in a:
    if "y" in name:
        continue
    print(name)


print()

a = ["brent", "taylor", "mitch", "jason", "buxby"]
for name in a:
    if "y" in name:
        continue
    print(name)   
else:
    print("No names with 'y' found") 
```
