### Define a Function (def)

```python
# This is a function definition

def print_greeting(name):
    print(f"Hello, {name}!")
print_greeting("Alice")
```

### Def: Combine First + Last Name

```python
def make_full_name(first_name, last_name):
    """Combine first and last name into a full name."""
    full_name = first_name + " " + last_name
    return full_name
make_full_name("Real", "Python")
```

Alternative (f-function)

```python
def make_full_name(first_name, last_name):
    print(f"{first_name} {last_name}")
make_full_name("Real", "Python")
```

### Def: Print on Two Lines

```python
def print_profile(name, language):
    name_line = "Name: " + name
    language_line = "Language: " + language
    print(name_line)
    print(language_line)
print_profile("Jean Baptiste", "French")
```

Alternative (f-function)

```python
def print_profile(name, language):
    print(f"Name: {name}")
    print(f"Language: {language}")
print_profile("Real Python", "Python")
```

### for Loops

```python
# Test with a list of items. Tuples (in parentheses) work the same way, but they are immutable (cannot be changed).
a = ["foo", "bar", "baz"]
for item in a:
    print(item)
```

```python
# Test with a dictionary of items.
a = {"foo": 1, "bar": 2, "baz": 3}
for item in a:
    print(item)

print()

# Test with a dictionary of items. Return values.
a = {"foo": 1, "bar": 2, "baz": 3}
for item in a.values():
    print(item)

print()

# Test with a dictionary of items. Return values.
a = {"foo": 1, "bar": 2, "baz": 3}
for k, n in a.items():
    print(f"key={k}, num={n}")
```

#### Range

```python
a = range(5)
for i in a:
    print(i)

print()

a = range(0,5)
for i in a:
    print(i)

print()

a = range(0,5,2)
for i in a:
    print(i)
```

#### Break & Continue & Else

```python
a = ["brent", "taylor", "mitch", "jason", "buxby"]
for name in a:
    if "y" in name:
        break 
    print(name)

print()

a = ["brent", "taylor", "mitch", "jason", "buxby"]
for name in a:
    if "y" in name:
        continue
    print(name)


print()

a = ["brent", "taylor", "mitch", "jason", "buxby"]
for name in a:
    if "y" in name:
        continue
    print(name)   
else:
    print("No names with 'y' found") 
```

### Conditional Statements

if, else, and elif let you run different code depending on whether conditions are true or false. Python checks each condition top to bottom and stops at the first one that's true — the rest are skipped. Only if is required; elif and else are optional.

```python
x = 10

y = 3

if x > y:
    print(f"{x} is greater than {y}")
```

```python
# Test with a list of items.
sports_list = ["soccer", "basketball", "tennis", "baseball"]

if "soccer" in sports_list:
    print("Soccer is in the list of sports.")
```

```python
# Test with user prompts and conditionals.
prompt = "How many calories did you consume? "
user_input = int(input(prompt))
print(f"Total calories consumed today: {user_input}")

if int(user_input) >= 1600:
    print("Don't eat it all!")

elif int(user_input) < 300:
    print("Eat a larger portion")

else:
    print("Sure, you can eat it!")
```

```python
# One liner statements
num_cals = 1000
print(f"Total calories consumed today: {num_cals}")

if (num_cals) >= 1600: print("Don't eat it all!")

elif (num_cals) < 300: print("Eat a larger portion")

else: print("Sure, you can eat it!")
```

```python
# Conditional Expressions: used as a way to assign variables based on results of a comparison.
dog_lover = True

buy_a_dog = "Yes" if dog_lover else "No"
print(buy_a_dog)
```
