```python
import pandas as pd
import os
```

```python
# Define file paths
ccg_file = '../merged/Hip Replacement CCG.csv'
provider_file = '../merged/Hip Replacement Provider.csv'

# Load the full dataframes
ccg_df = pd.read_csv(ccg_file, low_memory=False)
provider_df = pd.read_csv(provider_file, low_memory=False)
```

```python
# Print columns
print("Columns in CCG:")
print(ccg_df.columns.tolist())
print("\nColumns in Provider:")
print(provider_df.columns.tolist())
```

```python
# Check if columns are the same
ccg_cols = set(ccg_df.columns)
provider_cols = set(provider_df.columns)

if ccg_cols == provider_cols:
    print("Columns are identical.")
else:
    print("Columns differ.")
    print("Unique to CCG:", ccg_cols - provider_cols)
    print("Unique to Provider:", provider_cols - ccg_cols)
```

# Knee Replacement Column Comparison

Comparing columns between Knee Replacement CCG and Provider merged files.

```python
# Define file paths for Knee
knee_ccg_file = '../merged/Knee Replacement CCG.csv'
knee_provider_file = '../merged/Knee Replacement Provider.csv'

# Load the dataframes (only headers for efficiency)
knee_ccg_df = pd.read_csv(knee_ccg_file, nrows=0)
knee_provider_df = pd.read_csv(knee_provider_file, nrows=0)
```

```python
# Print columns for Knee
print("Columns in Knee CCG:")
print(knee_ccg_df.columns.tolist())
print("\nColumns in Knee Provider:")
print(knee_provider_df.columns.tolist())
```

```python
# Check if columns are the same for Knee
knee_ccg_cols = set(knee_ccg_df.columns)
knee_provider_cols = set(knee_provider_df.columns)

if knee_ccg_cols == knee_provider_cols:
    print("Knee columns are identical.")
else:
    print("Knee columns differ.")
    print("Unique to CCG:", knee_ccg_cols - knee_provider_cols)
    print("Unique to Provider:", knee_provider_cols - knee_ccg_cols)
```

```python
# Find duplicate rows except for Provider Code column
# Drop Provider Code and concatenate
ccg_no_provider = ccg_df.drop(columns=['Provider Code'])
provider_no_provider = provider_df.drop(columns=['Provider Code'])

# Add source column for identification
ccg_no_provider['Source'] = 'CCG'
provider_no_provider['Source'] = 'Provider'

# Concatenate
combined_df = pd.concat([ccg_no_provider, provider_no_provider], ignore_index=True)

# Find duplicates
duplicates = combined_df[combined_df.duplicated(keep=False)]

print(f"Number of duplicate rows (ignoring Provider Code): {len(duplicates)}")
if len(duplicates) > 0:
    print("Sample duplicates:")
    print(duplicates.head())
else:
    print("No duplicates found.")
```
