# Time Series Hip and Knee Replacement Data Merging

This notebook loads and merges the Time Series Hip and Knee Replacement data from 2016/17, 2017/18, and 2018/19.

```python
import pandas as pd
import os
```

```python
# Define file paths
files = {
    "1617": "../external/Time Series Hip and Knee Replacement 1617.csv",
    "1718": "../external/Time Series 1718.csv",
    "1819": "../external/Time Series Hip and Knee Replacement 1819.csv"
}

# Load dataframes with appropriate separators
df_1617 = pd.read_csv(files["1617"], sep=',')
df_1718 = pd.read_csv(files["1718"], sep=',')
df_1819 = pd.read_csv(files["1819"], sep=',')
```

```python
# Check columns
print("Columns in 1617:")
print(df_1617.columns.tolist())
print("\nColumns in 1718:")
print(df_1718.columns.tolist())
print("\nColumns in 1819:")
print(df_1819.columns.tolist())

# Check if columns are the same
cols_1617 = set(df_1617.columns)
cols_1718 = set(df_1718.columns)
cols_1819 = set(df_1819.columns)

if cols_1617 == cols_1718 == cols_1819:
    print("\nAll columns are the same.")
else:
    print("\nColumns differ:")
    print("Unique to 1617:", cols_1617 - cols_1718 - cols_1819)
    print("Unique to 1718:", cols_1718 - cols_1617 - cols_1819)
    print("Unique to 1819:", cols_1819 - cols_1617 - cols_1718)

# Check shapes
print("\nShape of 1617:", df_1617.shape)
print("Shape of 1718:", df_1718.shape)
print("Shape of 1819:", df_1819.shape)
```

```python
# Check data types
print("Dtypes of 1617:")
print(df_1617.dtypes)
print("\nDtypes of 1718:")
print(df_1718.dtypes)
print("\nDtypes of 1819:")
print(df_1819.dtypes)
```

```python
# Sample data
print("Head of 1617:")
print(df_1617.head())
print("\nHead of 1718:")
print(df_1718.head())
print("\nHead of 1819:")
print(df_1819.head())
```

```python
# Check for missing values
print("Missing values in 1617:")
print(df_1617.isnull().sum())
print("\nMissing values in 1718:")
print(df_1718.isnull().sum())
print("\nMissing values in 1819:")
print(df_1819.isnull().sum())
```

```python
# Since columns may differ, add year and concatenate
df_1617['CSVYear'] = '2016/17'
df_1718['CSVYear'] = '2017/18'
df_1819['CSVYear'] = '2018/19'

# Concatenate
merged_df = pd.concat([df_1617, df_1718, df_1819], ignore_index=True)
print("Merged shape:", merged_df.shape)
print("Merged columns:", list(merged_df.columns))
```

```python
# Check for duplicates
duplicates = merged_df.duplicated()
print("Number of duplicate rows:", duplicates.sum())
if duplicates.sum() > 0:
    print("Duplicate rows:")
    print(merged_df[duplicates].head())
```
