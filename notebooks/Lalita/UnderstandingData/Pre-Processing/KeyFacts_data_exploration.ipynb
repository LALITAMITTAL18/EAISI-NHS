# Data Exploration for Hip and Knee Replacement Files

This notebook explores the three CSV files to understand their structure and identify potential issues for merging.

Files:
- Key Facts Hip and Knee Replacement 1617.csv
- Key Facts Hip and Knee Replacements 1718.csv
- Key Facts Hip and Knee Replacements 1819.csv

```python
import pandas as pd
import os
```

```python
# Define file paths
file_1617 = '../external/Key Facts Hip and Knee Replacement 1617.csv'
file_1718 = '../external/Key Facts Hip and Knee Replacements 1718.csv'
file_1819 = '../external/Key Facts Hip and Knee Replacements 1819.csv'

# Load the dataframes
df_1617 = pd.read_csv(file_1617)
df_1718 = pd.read_csv(file_1718)
df_1819 = pd.read_csv(file_1819)
```

```python
# Check shapes
print("Shape of 1617:", df_1617.shape)
print("Shape of 1718:", df_1718.shape)
print("Shape of 1819:", df_1819.shape)
```

```python
# Check columns
print("Columns of 1617:", list(df_1617.columns))
print("Columns of 1718:", list(df_1718.columns))
print("Columns of 1819:", list(df_1819.columns))
```

```python
# Check data types
print("Dtypes of 1617:")
print(df_1617.dtypes)
print("\nDtypes of 1718:")
print(df_1718.dtypes)
print("\nDtypes of 1819:")
print(df_1819.dtypes)
```

```python
# Sample data
print("Head of 1617:")
print(df_1617.head())
print("\nHead of 1718:")
print(df_1718.head())
print("\nHead of 1819:")
print(df_1819.head())
```

```python
# Check for missing values
print("Missing values in 1617:")
print(df_1617.isnull().sum())
print("\nMissing values in 1718:")
print(df_1718.isnull().sum())
print("\nMissing values in 1819:")
print(df_1819.isnull().sum())
```

```python
# Check unique values in key columns
key_columns = ['Lookup', 'Procedure', 'Organisation Type', 'Organisation Code', 'Measure']

for col in key_columns:
    print(f"\nUnique in {col} for 1617: {df_1617[col].nunique()}")
    print(f"Unique in {col} for 1718: {df_1718[col].nunique()}")
    print(f"Unique in {col} for 1819: {df_1819[col].nunique()}")
```

```python
# Check if columns are identical
cols_1617 = set(df_1617.columns)
cols_1718 = set(df_1718.columns)
cols_1819 = set(df_1819.columns)

print("Columns identical 1617 and 1718:", cols_1617 == cols_1718)
print("Columns identical 1718 and 1819:", cols_1718 == cols_1819)
print("Columns identical 1617 and 1819:", cols_1617 == cols_1819)

if cols_1617 != cols_1718:
    print("Difference 1617-1718:", cols_1617.symmetric_difference(cols_1718))
if cols_1718 != cols_1819:
    print("Difference 1718-1819:", cols_1718.symmetric_difference(cols_1819))

# Add year column
df_1617['Year'] = '2016-17'
df_1718['Year'] = '2017-18'
df_1819['Year'] = '2018-19'

# Concatenate
merged_df = pd.concat([df_1617, df_1718, df_1819], ignore_index=True)
print("Merged shape:", merged_df.shape)
print("Merged columns:", list(merged_df.columns))
```

```python
# Save the merged dataframe to DATA/merged
import os
os.makedirs('../merged', exist_ok=True)
merged_df.to_csv('../merged/Key Facts Hip and Knee Replacement.csv', index=False)
print("Merged data saved to ../merged/merged_data.csv")
```

Unique value for lookup is changed because of merging the data. Hence need to add years to keep every Procedure, OrganizationType, Measure unique

CReate a Column Name with LookupYear and merge Year with lookup in merged dataframe 


```python
merged_df['LookupYear'] = merged_df['Lookup'] + '_' + merged_df['Year']

print(f"Shape of merged dataframe: {merged_df.shape}")

key_columns = ['LookupYear', 'Lookup', 'Procedure', 'Organisation Type', 'Organisation Code', 'Measure']

for col in key_columns:
    print(f"\nUnique in {col} for 1617: {merged_df[col].nunique()}")
```

```python
import matplotlib.pyplot as plt

# Ensure columns are numeric
merged_df['Improved'] = pd.to_numeric(merged_df['Improved'], errors='coerce')
merged_df['Unchanged'] = pd.to_numeric(merged_df['Unchanged'], errors='coerce')
merged_df['Worsened'] = pd.to_numeric(merged_df['Worsened'], errors='coerce')

# Group by Measure and sum the outcomes
outcomes_by_measure = merged_df.groupby('Measure')[['Improved', 'Unchanged', 'Worsened']].sum()

# Plot grouped bar chart with specified colors
outcomes_by_measure.plot(kind='bar', figsize=(12, 8), color=['green', 'yellow', 'red'])
plt.title('Total Cases by Measure and Outcome')
plt.xlabel('Measure')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.legend(title='Outcome')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by Procedure and sum the outcomes
outcomes_by_procedure = merged_df.groupby('Procedure')[['Improved', 'Unchanged', 'Worsened']].sum()

# Plot grouped bar chart with specified colors
outcomes_by_procedure.plot(kind='bar', figsize=(12, 8), color=['green', 'yellow', 'red'])
plt.title('Total Cases by Procedure and Outcome')
plt.xlabel('Procedure')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.legend(title='Outcome')
plt.show()
```
