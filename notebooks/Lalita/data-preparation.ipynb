```python
# pandas options
pd.set_option("display.max.columns", None)
pd.set_option("display.max.rows", None)
pd.set_option("display.precision", 2)
```

```python
# Load the Hip Replacement CCG data
file_path = r"c:\Users\mittall\source\EAISI\NHS\EAISI-NHS\notebooks\Lalita\UnderstandingData\merged\Hip Replacement CCG.csv"
df = pd.read_csv(file_path, sep=';')
```

```python
# Rename columns according to the specified rules
def rename_column(col):
    if col.startswith('Pre-Op Q '):
        # Replace 'Pre-Op Q ' with 't0_' and replace spaces with underscores
        new_col = col.replace('Pre-Op Q ', 't0_').replace(' ', '_')
        return new_col
    elif col.startswith('Hip Replacement Pre-Op Q '):
        # Replace 'Hip Replacement Pre-Op Q ' with 't0_HR_' and replace spaces with underscores
        new_col = col.replace('Hip Replacement Pre-Op Q ', 't0_HR_').replace(' ', '_')
        return new_col
    elif col.startswith('Post-Op_Q_'):
        # Replace 'Post-Op_Q_' with 't1_' 
        new_col = col.replace('Post-Op_Q_', 't1_', 1)
        return new_col
    elif col.startswith('Hip_Replacement_Post-Op_Q_'):
        # Replace 'Hip_Replacement_Post-Op_Q_' with 't1_HR_'
        new_col = col.replace('Hip_Replacement_Post-Op_Q_', 't1_HR_', 1)
        return new_col
    elif '_Post-Op_Q_' in col:
        # Handle special cases like Hip_Replacement_EQ5D_Index_Post-Op_Q_Predicted
        new_col = col.replace('_Post-Op_Q_', '_')
        if new_col.startswith('Hip_Replacement_'):
            new_col = 't1_HR_' + new_col[len('Hip_Replacement_'):]
        return new_col
    else:
        # For other columns, just replace spaces with underscores
        return col.replace(' ', '_')

# Apply the renaming
df.columns = [rename_column(col) for col in df.columns]

print("Columns renamed successfully.")
print("New column names:")
print(df.columns.tolist())
```

```python
# Create a comparison table of old vs new column names
# First, reload the data to get original column names
df_temp = pd.read_csv(file_path, sep=';', nrows=0)  # Only read headers
original_columns = df_temp.columns.tolist()

# Get the new columns
new_columns = df.columns.tolist()

# Create a comparison dataframe
comparison_df = pd.DataFrame({
    'Original_Column_Name': original_columns,
    'New_Column_Name': new_columns
})

print("Column Name Comparison Table:")
print(comparison_df.to_string(index=False))
```
